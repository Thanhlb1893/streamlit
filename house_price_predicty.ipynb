```python
import pandas as pd
import joblib
import numpy as np

# 1. Nạp lại mô hình và danh sách cột (Giữ nguyên đoạn này của bạn)
loaded_model = joblib.load('house_price_model.pkl')
model_cols = joblib.load('model_features.pkl')

print("Đã nạp mô hình thành công!")
print("Các thông tin cần nhập:", model_cols)

# 2. TẠO DỮ LIỆU GIẢ ĐỊNH (Thay vì dùng X_test_new)
# Nhập thông số cho căn nhà bạn muốn định giá
# Lưu ý: Tên cột phải khớp y hệt danh sách model_cols in ra ở trên
my_house_data = {
    'Age': [5],             # Nhà mới xây 5 năm
    'Total_Qua': [8],       # Chất lượng tốt (thang 10)
    '1st Flr SF': [1200],   # Diện tích tầng 1 (feet vuông)
    'Garage Area': [500]    # Diện tích Gara
}

# Chuyển thành bảng (DataFrame) để mô hình hiểu
test_house = pd.DataFrame(my_house_data)

# 3. Dự đoán
predicted_price_log = loaded_model.predict(test_house)
real_price = np.exp(predicted_price_log) # Mũ lên để ra tiền thật

print("-" * 30)
print(f"Dự đoán giá nhà: ${real_price[0]:,.2f}")
```
